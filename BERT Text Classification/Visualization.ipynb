```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
### Based on the data from the experiments
epochs = [2,3,10,20]
claims_accuracy = [99.4, 93.6, 80.8, 50.8]
claims_precision = [98.9, 93.6, 83.7, 50.4]
abstracts_accuracy = [98.4, 96.2, 97.8, 98.4]
abstracts_precision = [97.2, 97.3, 95.8, 97.6]
```

```python
plt.figure(figsize=(18, 12))
plt.plot(epochs, claims_accuracy, "-o", label='claims acc', color="red")
plt.plot(epochs, claims_precision, "-o", label='claims pre', color="orange")
plt.plot(epochs, abstracts_accuracy, "-o", label='abstracts acc', color="blue")
plt.plot(epochs, abstracts_precision, "-o", label='abstracts pre', color="skyblue")
plt.xlabel("Fine-tuning Epochs", fontsize='xx-large')
plt.ylabel("Accuracy and Precision (%)", fontsize='xx-large')
plt.legend(fontsize='xx-large')
plt.show()
```
